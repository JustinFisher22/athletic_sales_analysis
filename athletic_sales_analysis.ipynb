```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
sales_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
sales_2020
```

```python
# Display the 2021 sales DataFrame
sales_2021
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
sales_2020.info()
```

```python
# Check the 2021 sales data types.
sales_2021.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales_data = pd.concat([sales_2020, sales_2021], axis=0)
combined_sales_data.reset_index(drop=True, inplace=True)
combined_sales_data
```

```python
# Check if any values are null.
combined_sales_data.isnull()
```

```python
# Check the data type of each column
combined_sales_data.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales_data['invoice_date'] = pd.to_datetime(combined_sales_data['invoice_date'])
combined_sales_data
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales_data.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
groupby_products_sold_reg_state_city = combined_sales_data.groupby(['region', 'state', 'city']).agg({'units_sold':'sum'})

# Rename the sum to "Total_Products_Sold". 
groupby_products_sold_reg_state_city = groupby_products_sold_reg_state_city.rename(columns={'units_sold':'Total_Products_Sold'})

# Show the top 5 results.
groupby_products_sold_reg_state_city = groupby_products_sold_reg_state_city.sort_values(by='Total_Products_Sold', ascending=False)
groupby_products_sold_reg_state_city.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
products_sold_region_state_city = pd.pivot_table(combined_sales_data,
             index = ['region', 'state', 'city'],
             values = 'units_sold',
             aggfunc = 'sum')  

# Rename the "units_sold" column to "Total_Products_Sold"
products_sold_region_state_city = products_sold_region_state_city.rename(columns={"units_sold":"Total_Products_Sold"})

# Show the top 5 results.
products_sold_region_state_city = products_sold_region_state_city.sort_values(by='Total_Products_Sold', ascending=False)
products_sold_region_state_city.head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
total_sales_reg_state_city = combined_sales_data.groupby(['region', 'state', 'city']).agg({'total_sales': 'sum'})

# Rename the "total_sales" column to "Total Sales"
total_sales_reg_state_city = total_sales_reg_state_city.rename(columns={'total_sales':'Total Sales'})

# Show the top 5 results.
total_sales_reg_state_city = total_sales_reg_state_city.sort_values(by='Total Sales', ascending=False)
total_sales_reg_state_city.head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pt_total_sales_reg_state_city = pd.pivot_table(combined_sales_data,
                                               index = ['region', 'state', 'city'],
                                               values = 'total_sales',
                                               aggfunc = 'sum')

# Optional: Rename the "total_sales" column to "Total Sales"
pt_total_sales_reg_state_city = pt_total_sales_reg_state_city.rename(columns={'total_sales':'Total Sales'})

# Show the top 5 results.
pt_total_sales_reg_state_city = pt_total_sales_reg_state_city.sort_values(by= 'Total Sales', ascending=False)
pt_total_sales_reg_state_city.head(5)
```

### 4. Determine which Retailer had the Most Sales

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
invoice_date_table = pd.pivot_table(combined_sales_data,
                                    index = 'invoice_date',
                                    values = 'total_sales',
                                    aggfunc = 'sum')

# Optional: Rename the "total_sales" column to "Total Sales"
invoice_date_table = invoice_date_table.rename(columns={'total_sales': 'Total Sales'})

# Show the table.
invoice_date_table
```

```python
# Resetting the index for invoice_date
combined_sales_data.reset_index(inplace=True)
# Resample the pivot table into daily bins, and get the total sales for each day.
combined_sales_data.set_index('invoice_date', inplace=True)
daily_sales = combined_sales_data.resample('D').sum()['total_sales']

# Sort the resampled pivot table in descending order on "Total Sales".

sorted_sales = daily_sales.sort_values( ascending=False)
sorted_sales.head(5)
```
